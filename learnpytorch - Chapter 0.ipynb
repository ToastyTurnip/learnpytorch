### Matrix Multiplication
The shapes of the multiplicands must match in the inner layer!
The resulting shape is that of the outer dimensions

```python
tensor1 = torch.rand(3,5)
tensor2 = torch.rand(5,4)
```

```python
tensor3 = torch.matmul(tensor1, tensor2)
#torch.matmul can be shortened to torch.mm
tensor3
```

```python
tensor3.shape
```
