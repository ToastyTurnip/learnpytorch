#### Creating a dataset to sort (classify).


```python
from sklearn.datasets import make_circles

n_samples = 1000

# (WRONG! Initial impression) X and y are the x coordinates and y coordinates respectively for the samples which when graphed should make circles.
# Looks like (X1, X2) are the coordinates for the points (the feature) and y determines if it is from the inner or outer circle in this example (the label);
X, y = make_circles(n_samples=n_samples, noise= 0.03, random_state=42)

X[:5], y[:5]
```

#### Visualizing the data

```python
import pandas as pd
circles = pd.DataFrame({
    # Note ':' means copy everything in that dimension, a number means select that index in that specific dimension (dimension is denoted by place in the array)
    "X1" : X[:, 0],
    "X2" : X[:, 1],
    "y": y, 
})

circles.head(10)
```

##### NOTE: Accessing arrays within dataframes in numpy

```python
# Getting the count of the labels

# Note that dataframes in pandas convert string into variable member names of the parent dataframe, in this case we defined "y" : y so the "y" is now a member variable of circles
print("Y stuff: \n", circles.y.value_counts())

# Another example
# print("X1s: \n", circles.X1.value_counts())
```

```python
import matplotlib.pyplot as plt

def plot_2d_data(df):

    # get unique labels
    unique_labels = df.y.unique()
    label_dict = dict()

    # using boolean indexing, separate depending ont he label (value of y)
    for i in unique_labels:
        label_dict[str(i)] = df[df["y"] == i] 

    # plotting the points per label
    plt.figure(figsize=(10,10))
    
    colors = ["red", "green", "yellow", "blue", "orange"] 

    colors_index = 0
    for key in label_dict.keys():
        plt.scatter(label_dict[key].X1, label_dict[key].X2, s=2, c=colors[colors_index], label=key)
        colors_index += 1
    
    plt.legend(prop={"size": 14})

    

plot_2d_data(circles)
```
